```python
import math

while True:
    group_number = int(input("    1  3  0   : "))
    area = 0.0

    if group_number == 1:
        basis = float(input(" : "))
        height = float(input(" : "))
        area = 0.5 * basis * height

    elif group_number == 2:
        side1 = float(input("  : "))
        side2 = float(input("  : "))
        angle = float(input(" ,  : "))
        area = 0.5 * side1 * side2 * math.sin(angle)

    else:
        a = float(input("  : "))
        b = float(input("  : "))
        c = float(input("  : "))
        perimeter = (a + b + c) / 2
        area = math.sqrt(perimeter * (perimeter - a) * (perimeter - b) * (perimeter - c))

    print(f" : {area:.2f}  \n")
```

```python
N = int(input())
Sum = 0.0
cycle = 0

while cycle <= N:
    term = ((-1) ** cycle) * (1 / (2 ** cycle))
    Sum =  Sum + term
    cycle =  cycle  + 1

    print(f"S = {Sum:.4f}")
```
